```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
def create_dataset(n_samples):
    x = np.linspace(-1,1,n_samples)
    y = 0.1*x + x*x + x*x*x
    
    return x,y
    

    
```

```python
n_samples = 100
x , y = create_dataset(n_samples)
```

```python
plt.plot(x,y)
plt.plot(x,y,'ro')
plt.axis([-2,2,-2,4])
plt.show()
```

```python
plt.plot(x,y, 'r--', x,y**2, 'bs', x,y**3, 'g^')
plt.show()
```
