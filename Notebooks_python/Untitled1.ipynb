```python
import numpy as np

# Load the energy price data
energy_prices = np.loadtxt('C:/Users/prafu/Desktop/rL/a.csv', delimiter=',')

# Set hyperparameters
num_episodes = 1000
epsilon = 0.1
gamma = 0.9

# Initialize Q-values
num_prices = energy_prices.shape[0]
num_actions = 2  # 0: Do not use energy, 1: Use energy
Q = np.zeros((num_prices, num_actions))

# Reinforcement Learning algorithm
for episode in range(num_episodes):
    state = 0  # Start at the beginning of the energy price data

    for t in range(num_prices):
        # Choose action using epsilon-greedy policy
        if np.random.uniform(0, 1) < epsilon:
            action = np.random.randint(0, num_actions)
        else:
            action = np.argmax(Q[state, :])

        # Determine reward based on energy price
        price = energy_prices[state]
        if price < 50:
            reward = 1  # Use energy when price is low
        else:
            reward = -1  # Do not use energy when price is high

        # Update Q-value based on Q-learning update rule
        Q[state, action] += gamma * (reward + np.max(Q[state, :]) - Q[state, action])

        # Move to the next state
        state += 1

    # Reset the state to the beginning for the next episode
    state = 0

# Determine the optimal energy usage policy based on the learned Q-values
optimal_policy = np.argmax(Q, axis=1)

# Print the optimal energy usage policy
print("Optimal Energy Usage Policy:")
print(optimal_policy)
```
